# ##########################  71.546 ############################### # Sorting Values. Adding Columns: Majors w Most Potential vs Lowest risk

##################################################################################################

```python
clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
```

### We will add this column to our clean_df as "spread" with the .insert() method ###

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
```

```python
clean_df.insert(1, 'Spread', spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

### Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile. Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.

```python
highest_potential = clean_df.sort_values('Mid-Career 90th Percentile Salary', ascending = False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
highest_spread = clean_df.sort_values('Spread', ascending = False)
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
highest_spread = clean_df.sort_values('Mid-Career Median Salary', ascending = False)
highest_spread[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

# ####################### 71.548 ####################################                            Grouping and Pivoting with Pandas

# which category of degrees has the highest average salary? Is it STEM, Business or HASS (Humanities, Arts, and Social Science)? #

```python
print(df['Group'].value_counts())
```

```python
clean_df.groupby('Group').count()
```

```python
group_means = clean_df.groupby('Group').mean()
group_means.head()
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

```python
isinstance(group_means, pd.DataFrame)  #checking if group_means is a dataframe.
```

```python
highest_avg_salary = group_means.sort_values('Mid-Career Median Salary', ascending = False)
isinstance(highest_avg_salary, pd.DataFrame)
highest_avg_salary.head()
```
